```python
movies_detail2 = pd.DataFrame(columns=['Genre', 'Studio', 'Gross%US'])

included_genres = ['action', 'comedy', 'adventure', 'animation', 'drama']
included_studios = ['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox']

for index, row in movies_detail.iterrows():
    if (row.Genre in included_genres and row.Studio in included_studios):
        movies_detail2 = movies_detail2.append({'Genre': row['Genre'], 'Studio': row['Studio'], 'Gross%US': row['Gross%US']}, ignore_index=True)
```

```python
movies_detail2.head()
```

```python
movies_detail2.info()
```

```python
movies_detail2.Genre = movies_detail2.Genre.astype('category')
movies_detail2.Studio= movies_detail2.Studio.astype('category')
```

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27) #size of A4 paper
w = sns.boxplot(data=movies_detail2, x='Genre', y='Gross%US')
plt.title('Domestic Gross % By Genre', fontsize=35)
plt.ylabel('Gross % US', fontsize=25)
plt.xlabel('Genre', fontsize=25)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
# plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=1, prop={'size':20})
plt.show()
```

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27) #size of A4 paper
genre_order = ['action', 'comedy', 'adventure', 'animation', 'drama']
sns.boxplot(data=movies_detail2, x='Genre', y='Gross%US', color='lightgrey', order=genre_order, showfliers=False)
sns.stripplot(data=movies_detail2, x='Genre', y='Gross%US', order=genre_order, hue='Studio')
plt.title('Domestic Gross % By Genre', fontsize=35)
plt.ylabel('Gross % US', fontsize=25)
plt.xlabel('Genre', fontsize=25)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=1, prop={'size':20})
plt.show()
```
